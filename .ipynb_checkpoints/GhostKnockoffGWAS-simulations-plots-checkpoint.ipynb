# Simulation studies using PAISA and UKB data

PAISA:
+ We have $n=11221$ samples from GLIMPSE2 output, but we kept only 8666 samples for some of the Paisa simulations
+ SNPs are filtered to those present on the GSA array chip with MAF > 0.01. This leaves ~250k SNPs.
+ Many samples are highly related. 

UKB:
+ We have ~300,000 British samples where related samples are all removed

```julia
using Plots
using ProgressMeter
using CSV
using DataFrames
using SnpArrays
using Statistics

global titlefontsize=10
global legendfontsize=7
global labelfontsize=10
global annotationfontsize=10
global guidefontsize=10

# function make_plot(power::Vector{Float64}, obs_fdr::Vector{Float64}, tgt_fdr::Vector{Float64})
#     power_plt = plot(tgt_fdr, power; xtick=false, ylim=(0, 1.0), ylabel="Power",
#         legend=false, left_margin = 5Plots.mm, titlefontsize=titlefontsize,
#         guidefontsize=guidefontsize, labelfontsize=labelfontsize, w=3)
#     fdr_plt = plot(tgt_fdr, obs_fdr; xlabel="target FDR", ylim=(0, 0.25), ylabel="obs FDR", legend=false, 
#         xgrid=false, left_margin = 5Plots.mm, guidefontsize=guidefontsize,
#         labelfontsize=labelfontsize, w=3)
#     Plots.abline!(fdr_plt, 1, 0, alpha=0.3, w=2, linestyle=:dot, color=:black)
#     return plot(power_plt, fdr_plt, layout = grid(2, 1, heights=[0.7 ,0.3]), fmt=:png)
# end

# on array chr10 SNPs
function summarize_paisa_result(seeds, tgt_fdr::Vector{Float64})
    powers = zeros(length(tgt_fdr))
    fdr = zeros(length(tgt_fdr))
    successes = 0
    @showprogress for seed in seeds
        try
            GK_df = CSV.read("/oak/stanford/groups/zihuai/paisa/sims/sim$seed.lmm.txt", DataFrame)
            qvals = GK_df[!, "qvals"]
            beta = GK_df[!, "true_beta"]
            GK_groups = GK_df[!, "group"]
            causal_groups = unique!(GK_groups[findall(!iszero, beta)])
            for (i, threshold) in enumerate(tgt_fdr)
                GK_discover_snps = findall(x -> x < threshold, qvals)
                GK_discover_groups = unique!(GK_groups[GK_discover_snps])
                GK_group_power = length(intersect(GK_discover_groups, causal_groups)) / length(causal_groups)
                GK_group_FDR = length(setdiff(GK_discover_groups, causal_groups)) / max(1, length(GK_discover_groups))
                powers[i] += GK_group_power
                fdr[i] += GK_group_FDR
            end
            successes += 1
        catch
            continue
        end
    end
    @show successes
    powers ./= successes
    fdr ./= successes
    return powers, fdr
end

# Power/FDR for 56k imputed SNPs (using Sue's independent loci's to compute FDR)
function summarize_paisa2_result(seeds, tgt_fdr::Vector{Float64})
    # read independent SNP list
    independent_snp_file = "/oak/stanford/groups/zihuai/paisa/sue/SNPs.In.GWAS.Loci.processed.txt"
    independent_snps = CSV.read(independent_snp_file, DataFrame)
    
    powers = zeros(length(tgt_fdr))
    fdr = zeros(length(tgt_fdr))
    successes = 0
    @showprogress for seed in seeds
#         try
            GK_df = CSV.read("/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt", DataFrame)
            qvals = GK_df[!, "qvals"]
            beta = GK_df[!, "true_beta"]            
            causal_loci = @views get_causal_loci(GK_df[findall(!iszero, beta), :], independent_snps)
            for (i, threshold) in enumerate(tgt_fdr)
                GK_discovered = @views get_causal_loci(GK_df[findall(x -> x < threshold, qvals), :], independent_snps)
                GK_group_power = length(intersect(GK_discovered, causal_loci)) / length(causal_loci)
                GK_group_FDR = length(setdiff(GK_discovered, causal_loci)) / max(1, length(GK_discovered))
                powers[i] += GK_group_power
                fdr[i] += GK_group_FDR
            end
            successes += 1
#         catch
#             continue
#         end
    end
    @show successes
    powers ./= successes
    fdr ./= successes
    return powers, fdr
end

# Power/FDR for 56k imputed SNPs (using knockoff groups to compute FDR)
function summarize_paisa4_result(seeds, tgt_fdr::Vector{Float64})
    powers = zeros(length(tgt_fdr))
    fdr = zeros(length(tgt_fdr))
    successes = 0
    @showprogress for seed in seeds
        try
            GK_df = CSV.read("/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt", DataFrame)
            qvals = GK_df[!, "qvals"]
            beta = GK_df[!, "true_beta"]
            GK_groups = GK_df[!, "group"]
            causal_groups = unique!(GK_groups[findall(!iszero, beta)])
            for (i, threshold) in enumerate(tgt_fdr)
                GK_discover_snps = findall(x -> x < threshold, qvals)
                GK_discover_groups = unique!(GK_groups[GK_discover_snps])
                GK_group_power = length(intersect(GK_discover_groups, causal_groups)) / length(causal_groups)
                GK_group_FDR = length(setdiff(GK_discover_groups, causal_groups)) / max(1, length(GK_discover_groups))
                powers[i] += GK_group_power
                fdr[i] += GK_group_FDR
            end
            successes += 1
        catch
            continue
        end
    end
    powers ./= successes
    fdr ./= successes
    return powers, fdr
end

# computes FWER on 56K SNPs using knockoff groups as true/false positives
function summarize_paisa_result5(seeds; marginal_cutoff = 0.05 / 692)
    FWER = 0
    FDP = 0.0
    power = 0.0
    successes = 0
    @showprogress for seed in seeds
        try
            GK_df = CSV.read("/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt", DataFrame)
            pvals = GK_df[!, "pvals"]
            beta = GK_df[!, "true_beta"]
            GK_groups = GK_df[!, "group"]
            causal_groups = unique!(GK_groups[findall(!iszero, beta)])
            
            discover_snps = findall(x -> x < marginal_cutoff, pvals)
            discover_groups = unique!(GK_groups[discover_snps])
            power += length(intersect(discover_groups, causal_groups)) / length(causal_groups)
            if length(setdiff(discover_groups, causal_groups)) > 0
                FWER += 1
            end
            
            FP = length(setdiff(discover_groups, causal_groups))
            FDP += FP / max(1, length(discover_groups))
            
            successes += 1
        catch
            continue
        end
    end
    @show successes
    power /= successes
    FWER /= successes
    FDP /= successes
    return power, FWER, FDP
end

# Marginal power/FDR for 56k imputed SNPs using the locis defined by Esteban (this is "Causal D-loci")
function summarize_paisa3_result(seeds, tgt_fdr::Vector{Float64}; marginal_cutoff=0.05/692)
    # read independent SNP list
    independent_snp_file = "/oak/stanford/groups/zihuai/paisa/sue/SNPs.In.GWAS.Loci.processed.txt"
    independent_snps = CSV.read(independent_snp_file, DataFrame)
    
    FDR = 0.0
    FWER = 0.0
    avg_power = 0.0

    successes = 0
    @showprogress for seed in seeds
        try
            outfile = "/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt"
            power, FP, FDP = count_false_discoveries(outfile, independent_snps, marginal_cutoff=marginal_cutoff)
            FDR += FDP
            avg_power += power
            if FP > 0
                FWER += 1
            end
            successes += 1
        catch
            continue
        end
    end
    FWER /= successes
    avg_power /= successes
    FDR /= successes
    return avg_power, FWER, FDR
end

function count_false_discoveries(
    outfile::String,
    independent_snps::DataFrame; # needs :rsid, :chr, :pos_hg$(hg_build), :pvals, :W, :group
    marginal_cutoff = 0.05 / 692
    )
#     outfile = "/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt"
    result = CSV.read(outfile, DataFrame)

    # get discovered locis 
    discovered = result[findall(x -> x < marginal_cutoff, result.pvals), :]
    augment_loci!(discovered, independent_snps)
    discovered_loci = unique(discovered.loci)
#     discovered_loci = get_causal_loci(discovered, independent_snps)
    
    # get true locis
    true_df = result[findall(!iszero, result.true_beta), :]
    augment_loci!(true_df, independent_snps)
    true_loci = unique(true_df.loci)
#     true_loci = get_causal_loci(true_df, independent_snps)

    # count power and number of false discoveries
    power = length(intersect(discovered_loci, true_loci)) / length(true_loci)
    FP = length(setdiff(discovered_loci, true_loci))
    FDP = FP / max(1, length(discovered_loci))

    return power, FP, FDP
end

function get_causal_loci(discovered::AbstractDataFrame, independent_snps::DataFrame)
    chr, pos, ref, alt = discovered.chr, discovered.pos_hg38, discovered.ref, discovered.alt
    locus = String[]
    for (c, p, r, a) in zip(chr, pos, ref, alt)
        idx = findfirst(
            x -> x.CHR == c && x.POS == p && x.REF == r && x.ALT == a, eachrow(independent_snps)
        )
        # consider potention ref <--> alt flip
        if isnothing(idx)
            idx = findfirst(
                x -> x.CHR == c && x.POS == p && x.REF == a && x.ALT == r, eachrow(independent_snps)
            )
            isnothing(idx) && error("a discovered SNP is not in `independent_snps`! Shouldn't happen")
        end
        push!(locus, independent_snps[idx, "clump_ID"])
    end
    return unique!(locus)
end

function augment_loci!(discovered, independent_snps)
    discovered[!, "loci"] = ["" for _ in 1:size(discovered, 1)]
    for row in eachrow(discovered)
        c, p, r, a = row.chr, row.pos_hg38, row.ref, row.alt
        idx = findfirst(
            x -> x.CHR == c && x.POS == p && x.REF == r && x.ALT == a, eachrow(independent_snps)
        )
        # consider potention ref <--> alt flip
        if isnothing(idx)
            idx = findfirst(
                x -> x.CHR == c && x.POS == p && x.REF == a && x.ALT == r, eachrow(independent_snps)
            )
            isnothing(idx) && error("a discovered SNP is not in `independent_snps`! Shouldn't happen")
        end
        row.loci = independent_snps[idx, "clump_ID"]
    end
    discovered
end

function summarize_UKB_result(seeds, tgt_fdr::Vector{Float64})
    powers = zeros(length(tgt_fdr))
    fdr = zeros(length(tgt_fdr))
    successes = 0
    @showprogress for seed in seeds
        try
            GK_df = CSV.read("/oak/stanford/groups/zihuai/solveblock/sims/sim$seed.txt", DataFrame)
            qvals = GK_df[!, "qvals"]
            beta = GK_df[!, "true_beta"]
            GK_groups = GK_df[!, "group"]
            causal_groups = unique!(GK_groups[findall(!iszero, beta)])
            for (i, threshold) in enumerate(tgt_fdr)
                GK_discover_snps = findall(x -> x < threshold, qvals)
                GK_discover_groups = unique!(GK_groups[GK_discover_snps])
                GK_group_power = length(intersect(GK_discover_groups, causal_groups)) / length(causal_groups)
                GK_group_FDR = length(setdiff(GK_discover_groups, causal_groups)) / max(1, length(GK_discover_groups))

                powers[i] += GK_group_power
                fdr[i] += GK_group_FDR
            end
            successes += 1
        catch
            continue
        end
    end
    @show successes
    powers ./= successes
    fdr ./= successes
    return powers, fdr
end

function summarize_result(pop, seeds, tgt_fdr::Vector{Float64})
    powers = zeros(length(tgt_fdr))
    fdr = zeros(length(tgt_fdr))
    successes = 0
    lmm = pop == "british" ? "" : "lmm."
    @showprogress for seed in seeds
        try
            GK_df = CSV.read("/oak/stanford/groups/zihuai/solveblock/sims_$pop/sim$seed.$(lmm)txt", DataFrame)
            qvals = GK_df[!, "qvals"]
            beta = GK_df[!, "true_beta"]
            GK_groups = GK_df[!, "group"]
            causal_groups = unique!(GK_groups[findall(!iszero, beta)])
            for (i, threshold) in enumerate(tgt_fdr)
                GK_discover_snps = findall(x -> x < threshold, qvals)
                GK_discover_groups = unique!(GK_groups[GK_discover_snps])
                GK_group_power = length(intersect(GK_discover_groups, causal_groups)) / length(causal_groups)
                GK_group_FDR = length(setdiff(GK_discover_groups, causal_groups)) / max(1, length(GK_discover_groups))
                powers[i] += GK_group_power
                fdr[i] += GK_group_FDR
            end
            successes += 1
        catch
            continue
        end
    end
    @show successes
    powers ./= successes
    fdr ./= successes
    return powers, fdr
end

function marginal_assoc(
    outfile::String;
    marginal_cutoff = 0.05 / 692,
    xdata = SnpData("/oak/stanford/groups/zihuai/paisa/GLIMPSE2_wave2_QC/snps_in_gwas/allchr.maf0.01"),
    x = xdata.snparray,
    )
    seed = split(basename(outfile), '.')[1][4:end]

    # simulation result
#     outfile = "/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt"
    result = CSV.read(outfile, DataFrame)
    result.idx = collect(1:size(result, 1))
    
    # get discovered
    discovered = result[findall(x -> x < marginal_cutoff, result.pvals), :]
    
    # true SNPs
    causal_idx = findall(!iszero, result.true_beta)
    true_df = result[causal_idx, :]
    x_idx = Int[]
    for row in eachrow(true_df)
        push!(x_idx, find_idx(row, xdata.snp_info))
    end
    xtrue = convert(Matrix{Float64}, @view(x[:, x_idx]), impute=true, center=true, scale=true)

    # check if discovered loci is a "true discovery" according to definition
    FP = 0
    found = falses(length(causal_idx))
    for row in eachrow(discovered)
        idx = find_idx(row, xdata.snp_info)
        snp = convert(Vector{Float64}, @view(x[:, idx]), impute=true, center=true, scale=true)
        corrs = abs.(cor(xtrue, snp))
        if all(corrs .<= 0.25)
            FP += 1
        end
        found[findall(corrs .>= 0.25)] .= true
    end
    power = sum(found) / length(causal_idx)

    # also compute FDP = (# FP) / (# discoveries)
    
    return power, FP
end

function find_idx(row::DataFrameRow, target::DataFrame)
    # unique match
    idxs = findall(x -> x == row.pos_hg38, target.position)
    if length(idxs) == 1
        return idxs[1]
    end
    
    # non unique match: search chr/ref/alt
    c, p, r, a = row.chr, row.pos_hg38, row.ref, row.alt    
    idx = findfirst(
        x -> x.chromosome == string(c) && x.position == p && x.allele1 == r && x.allele2 == a, eachrow(target)
    )
    # consider potention ref <--> alt flip
    if isnothing(idx)
        idx = findfirst(
            x -> x.chromosome == string(c) && x.position == p && x.allele1 == a && x.allele2 == r, eachrow(target)
        )
        isnothing(idx) && error("a discovered SNP is not in `independent_snps`! Shouldn't happen")
    end
    return idx
end

function summarize_paisa6_result(seeds; marginal_cutoff = 0.05 / 692)
    FWER = 0.0
    powers = 0.0
    @showprogress for seed in seeds
        outfile = "/oak/stanford/groups/zihuai/paisa/8666sims/sim$seed.lmm.txt"
        power, FP = marginal_assoc(outfile, marginal_cutoff=marginal_cutoff)
        if FP > 0
            FWER += 1
        end
        powers += power
    end
    FWER /= length(seeds)
    powers /= length(seeds)
    return powers, FWER
end
```
